# Introduction to Data Visualization with Seaborn

In this chapter, you will create and customize plots that visualize the relationship between two quantitative variables. To do this, you will use scatter plots and line plots to explore how the level of air pollution in a city changes over the course of a day and how horsepower relates to fuel efficiency in cars. You will also see another big advantage of using Seaborn - the ability to easily create subplots in a single figure!

## Visualizing Two Quantitative Variables

### scatterplot() vs. relplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
student_data = pd.read_csv('../datasets/student-alcohol-consumption.csv')
student_data.head()
```

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter")

# Show plot
plt.show()
```

```python
# Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
           col="study_time")

# Show plot
plt.show()
```

```python
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            row="study_time")
plt.show()
```

```python
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter")

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])

# Show plot
plt.show()
```

# Customizing scatter plots

```python
## Tamanho do ponto 

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

mpg =pd.read_csv('../datasets/mpg.csv')

mpg.head()
```

## Tamanho do ponto 

```python
# tamanho do ponto 
sns.relplot(x="horsepower", y="mpg", data= mpg, size="cylinders")

plt.show()
```

## Pontos coloridos

```python
# pontos coloridos
# usando hue
sns.relplot(x="horsepower", y="mpg", data= mpg, size="cylinders", hue='cylinders')

plt.show()
```

## Formato do ponto para representar 

```python
# pontos coloridos e formato para representaÃ§Ã£o 
sns.relplot(x="acceleration", y="mpg", data= mpg, hue='origin', style='origin')

plt.show()
```

```python
# tranaparencia de ponto
# usaremos alpha 
sns.relplot(x="acceleration", y="mpg", data= mpg, hue='origin', style='origin', alpha=0.4)

plt.show()
```

# Tipos de linha 

## linha

```python
sns.relplot( data=mpg, x='model_year', y='mpg', kind='line')

plt.show()
```

## Com representaÃ§Ã£o do desvio padÃ£o 

para isso estamos usando **ci='sd'**

```python
# Create line plot
sns.relplot( data=mpg, x='model_year', y='mpg', kind='line', ci='sd')

# Show plot
plt.show()
```

```python
# Sem a formataÃ§Ã£o 
sns.relplot( data=mpg, x='model_year', y='mpg', kind='line', ci=None)

# Show plot
plt.show()
```

## Dashes

```python
# dashes False
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", dashes=False)

# Show plot
plt.show()
```

```python
# Dashes True
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", dashes=True)

# Show plot
plt.show()
```

## Markers

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", dashes=False, markers=True)

# Show plot
plt.show()
```
