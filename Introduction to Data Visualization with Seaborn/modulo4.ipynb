# Introduction to Data Visualization with Seaborn

## Customizing Seaborn Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
survey_data =  pd.read_csv('../datasets/young-people-survey-responses.csv')
survey_data.head()
```

## Estilo "whitegrid"

```python
survey_data['Internet usage'].value_counts()
```

```python
survey_data['Internet usage'].value_counts().index
```

```python
# Set the style to "whitegrid"
sns.set_style("whitegrid")

# Create a count plot of survey responses
category_order = survey_data['Internet usage'].value_counts().index

sns.catplot(x='Internet usage', 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

## Palheta de cores 

```python
# Set the style to "whitegrid"
sns.set_style("whitegrid")

sns.set_palette("Purples")

# Create a count plot of survey responses
category_order = survey_data['Internet usage'].value_counts().index

sns.catplot(x='Internet usage', 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
# Set the style to "whitegrid"
sns.set_style("whitegrid")

sns.set_palette("RdBu")

# Create a count plot of survey responses
category_order = survey_data['Internet usage'].value_counts().index

sns.catplot(x='Internet usage', 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

## Criando proprio palheta de  cores 

```python
custom_palette = ['#FBB4AE', '#B3CDE3', '#CCEBC5',
'#DECBE4', '#FED9A6', '#FFFFCC',
'#E5D8BD', '#FDDAEC', '#F2F2F2']
sns.set_palette(custom_palette)

# Set the context to "talk"
sns.set_context("talk")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

## Adicionando titulo 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

mpg =pd.read_csv('../datasets/mpg.csv')

mpg.head()
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Show plot
plt.show()
```

## Colocando nome dos eixos 

```python
# Create line plot
sns.set_context('paper')
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year", ylabel="Average MPG")


# Show plot
plt.show()
```

## Rotacionando o texto do eixo x

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

survey_data =  pd.read_csv('../datasets/young-people-survey-responses.csv')

# Set palette to "Blues"
sns.set_palette("Blues")

# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender", y="Age", data=survey_data,  kind="box")

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not")

# Show plot
plt.show()
```

```python
# Set the figure style to "dark"
sns.set_style("dark")

# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Finances", 
                data=survey_data, kind="bar",
                col="Gender")

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
       ylabel="% Who Like Techno")

# Show plot
plt.show()
```
