```python
import matplotlib.pyplot as plt
import pandas as pd
import calendar
```

```python
df =  pd.read_csv('../datasets/seattle_weather.csv')
```

```python
df.head()
```

```python
seattle_weather = df[:12] # so queremos trabalhar com dado de um ano
seattle_weather.shape
```

```python
# DATE e do tipo inteiro que representa o mes, vamo criar uma coluna do tipo mes 
seattle_weather['MONTH'] = seattle_weather['DATE'].apply(lambda x: pd.Timestamp(month=x, day=1, year=2000).strftime('%b'))
seattle_weather['MONTH']
```

```python
# fazendo analogo 
df =  pd.read_csv('../datasets/austin_weather.csv')
austin_weather = df[:12] # so queremos trabalhar com dado de um ano
austin_weather['MONTH'] = austin_weather['DATE'].apply(lambda x: pd.Timestamp(month=x, day=1, year=2000).strftime('%b'))
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

## Adding markers

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],
seattle_weather["MLY-TAVG-NORMAL"],
marker="v")
plt.show()
```

## Estilo de linhas 


```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],
seattle_weather["MLY-TAVG-NORMAL"],
marker="v", linestyle="--")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],
seattle_weather["MLY-TAVG-NORMAL"],
marker="v", linestyle="None")
plt.show()
```

## Cor da linha 

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],
seattle_weather["MLY-TAVG-NORMAL"],
marker="v", linestyle="--", color="r")
plt.show()
```

## Nomes do eixo e titutlo 

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],
seattle_weather["MLY-TAVG-NORMAL"],
marker="v", linestyle="--", color="r")

ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
ax.set_title("Weather in Seattle")

plt.show()
```

# Small multiples

No proximo exemplo, o grafico esta confuso pois existe varios grafico juntos, a ideia Ã© fazer varias telas para organizar melhor

```python
fig, ax = plt.subplots()
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
color='r')
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='r')
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='r')
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='b')
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='b')
plt.show()
```

```python
# 3 linha e duda coluna de tela
fig, ax = plt.subplots(3, 2)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],
color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='b')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
color='r')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='r')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='r')
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
```

**Obs1:**  quando estamos lidando com uma coluna podemos usar como exemplo de cima, mas for mais remos que usar a notaÃ§Ã£o "az[1,3]"  por exemplo

**OBS2:** Repare que a coluna y esta de forma independente 

## Deixando o eixo y com o mesmo tamanho

```python
# repare nesse grafico e do anterior principalemnte em relaÃ§Ã£o ao eixo y
# quem faz toda essa magia Ã© sharey=True
# da linha fig, ax = plt.subplots(2, 1, sharey=True)
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],
color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='b')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
color='r')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"],
linestyle='--', color='r')
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"],
linestyle='--', color='r')
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
```
